## Text Summarization -  Extractive Method

Given a large text of information, we should summarize the large text and extract the meaning of the text.

Extractive Method - Takes key phrases from source document and combines them to create a summary

```python
#IMPORT LIBRARIES 

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
#CALUCLATE THE FREQUENCY OF THE WORDS IN THE TEXT

def frequency_table(text_string):
    stopWords = set(stopwords.words("english"))
    words = word_tokenize(text_string)
    ps = PorterStemmer()
    
    freqTable = dict()
    for word in words:
        word = ps.stem(word)
        if word in stopWords:
            continue
        if word in freqTable:
            freqTable[word] += 1
        else:
            freqTable[word] = 1

    return freqTable
```

```python
mywordfreq = frequency_table(text)
mywordfreq
```

```python
len(mywordfreq)
```

```python
#CALCULATE SCORE FOR EVERY SENTENCE

def score_sentences(sentences,freqTable):
  sent_value = dict()

  for sentence in sentences:
    word_count = (len(word_tokenize(sentence)))
    word_count_except_sw = 0

    for wordValue in freqTable:
      if wordValue in sentence.lower():
        word_count_except_sw += 1
        if sentence[:] in sent_value:
          sent_value[sentence[:]] += freqTable[wordValue]
        else:
          sent_value[sentence[:]] = freqTable[wordValue]

    if sentence[:] in sent_value:
       sent_value[sentence[:]] = sent_value[sentence[:]] / word_count_except_sw 

  return sent_value        
```

```python
mysentences = sent_tokenize(text)
mysentscores = score_sentences(mysentences,mywordfreq)
mysentscores
```

#Since long sentences are little hard to read as it is clumpsy and occupies more memory, we will modify the user-defined function of score_sentences.

#But, long sentences will have advantage over small sentences.

```python
def score_sentences(sentences,freqTable):
  sent_value = dict()

  for sentence in sentences:
    word_count = (len(word_tokenize(sentence)))
    word_count_except_sw = 0

    for wordValue in freqTable:
      if wordValue in sentence.lower():
        word_count_except_sw += 1
        if sentence[0:15] in sent_value:
          sent_value[sentence[0:15]] += freqTable[wordValue]
        else:
          sent_value[sentence[0:15]] = freqTable[wordValue]

    if sentence[0:15] in sent_value:
       sent_value[sentence[0:15]] = sent_value[sentence[0:15]] / word_count_except_sw 

  return sent_value
```

```python
mysentences = sent_tokenize(text)
mysentscores = score_sentences(mysentences,mywordfreq)
mysentscores
```

```python
len(mysentscores.keys())
```

```python
#Find the average score

def find_average_score(sent_value):
  sumValues = 0
  for entry in sent_value:
    sumValues += sent_value[entry]

  average = (sumValues/len(sent_value))

  return average
```

```python
myavg = find_average_score(mysentscores)
myavg
```

```python
#CREATE A FUCNTION  TO GENERATE SUMMARY

def generate_summary(sentences, sent_value, threshold):
  sentence_count = 0
  summary = ''

  for sentence in sentences:
    if sentence[0:15] in sent_value and sent_value[sentence[0:15]] >= (threshold):
      summary += " " + sentence
      sentence_count += 1

  return summary
```

```python
mysummary = generate_summary(mysentences,mysentscores, myavg)
print(len(mysummary))
mysummary
```

**Creating a function to call all the above functions** 

```python
def summarization(text,threshold):

  #1)Create the word frequency table
  freq_table = frequency_table(text)

  #2) Tokenize the sentences
  sentences = sent_tokenize(text)

  #3) Important algorithm: score the sentences 
  sentence_scores = score_sentences(sentences, freq_table)

  #4) Find the threshold
  threshold = find_average_score(sentence_scores)

  #5) Important algorithm: Generate the summary
  summary = generate_summary(sentences, sentence_scores, threshold)

  return summary
```

```python
result = summarization(text,2)
print(len(result))
result
```
